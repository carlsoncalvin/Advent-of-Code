## Day 1 part 1

I used floor instead of mod because I didn't think about it until after I had figured it out using floor.... so I kept it.

```python
%%time

direction = {"R": 1, "L": -1}
loc = 50
count = 0

with open("day1.txt", "r") as rotations:
    for line in rotations:
        rot = direction[line.strip()[0]]
        val = int(line.strip()[1:])
        
        new_loc = loc + rot * val
        
        if new_loc > 99:
            new_loc -= new_loc//100 * 100
        elif new_loc < 0:
            new_loc += -(new_loc//100 + 1) * 100 + 100
        
        if new_loc == 0:
            count += 1
        
        loc = new_loc

print(count)
```

## Day 1 part 2

This took an embarassingly long amount of time to get right. I made a test version of the script and ran through a few examples until I figured out that I needed two exceptions for negative rotations.

```python
%%time

direction = {"R": 1, "L": -1}
loc = 50
count = 0

with open("day1.txt", "r") as rotations:
    for line in rotations:
        line = line.strip()
        
        rot = direction[line[0]]
        val = int(line[1:])
        
        new_loc = loc + rot * val
        
        if new_loc > 99:
            count += new_loc//100
            new_loc -= new_loc//100 * 100
        elif new_loc < 0:
            if new_loc % 100 == 0:
                count += 1
            if loc == 0:
                count -= 1
            count += -(new_loc//100)
            new_loc = new_loc%100
        elif new_loc == 0:
            count += 1
        
        loc = new_loc

print(count)
```

## Day 2 part 1

```python
%%time

with open("day2.txt", "r") as file:
    ranges = file.read()

ranges = ranges.split(",")

id_sum = 0

for num_range in ranges:
    start, end = num_range.split("-")

    values = np.arange(int(start), int(end) + 1)

    for value in values:
        value = str(value)
        
        if len(value) % 2 != 0:
            continue
        
        i = int(len(value) / 2)
        
        if value[i:] == value[:i]:
            id_sum += int(value)
                
print(id_sum)
```

```python
%%time

with open("day2.txt", "r") as file:
    ranges = file.read()

ranges = ranges.split(",")

id_sum = 0

for num_range in ranges:
    start, end = num_range.split("-")

    values = np.arange(int(start), int(end) + 1)

    for value in values:
        value = str(value)
        n_start = len(value) // 2

        for n in range(n_start, -1, -1):
            if n == 0 or len(value) % n != 0:
                continue
            
            split = [value[i:i+n] for i in range(0, len(value), n)]
            
            if split.count(split[0]) == len(split):
                id_sum += int(value)
                break
        
                
print(id_sum)
```

## Day 3 part 1

```python
%%time

joltage = 0
with open("day3.txt", "r") as file:
    for line in file:
        line = line.strip()
        
        val_1 = max(line[:-1])
        idx = line.find(val_1)
        
        val_2 = max(line[idx+1:])
        
        joltage += int(val_1 + val_2)

print(joltage)
```

## Day 3 part 2

```python
%%time

joltage = 0
with open("day3.txt", "r") as file:
    for line in file:
        line = line.strip()
        n = len(line)
        
        jolts= max(line[:-11])
        i = line.find(jolts)
        
        # check for trivial case
        if i == n - 12:
            joltage += int(line[-12:])
            continue
        
        while i < n:
            end_idx = -11+len(jolts)
            
            if end_idx == 0:
                end_idx = None
            
            new_val = max(line[i+1:end_idx])
            jolts += new_val
            
            i = line[i+1:].find(new_val) + i + 1
            
            if len(jolts) == 12:
                break


        joltage += int(jolts)

print(joltage)
```
